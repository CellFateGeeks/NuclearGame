# Nuclear Game (NG) - Analysis

### Import modules

```python
import ng_tools as ngt
```

### Path to JSON file 
Path to compressed JSON file (output of NG-Segmentation)

```python
path = "/Volumes/emilio_passport/nuclei/data/d6Ngn2/out_ng/output.json.gz"
```

## Start NG - Analysis

```python
nga = ngt.NuclearGame_Analysis(path)
```

We can access raw data as a DataFrame:

```python
nga.df_raw
```

We can print the measured nuclear features:

```python
nga.print_features()
```

### Select cells of interest from Scatter Plot
Using the tool **Lasso Select** (found at the upper right of the plot) select the cells of interest. Modify *feature1* and *feature2* for changing the features plotted. 

```python
f, t = nga.scatter_widget(feature1 = 'avg_intensity_dapi', 
                         feature2 = 'nuclear_area', 
                         df = None, # If None is given, then uses raw data (nga.df_raw). 
                         xlog = True, 
                         ylog = True
                        )
f
```

Save selection as DataFrame:

```python
nga.df_filtered = ngt.selection2df(table = t)
nga.df_filtered
```

We can plot any of the features measured by changing the parameter 'feature'.

```python
fig = nga.plot_boxplot_hist(feature = "avg_intensity_b3", 
                            df = nga.df_filtered # If None is given, then uses raw data (nga.df_raw)
                           )
fig.show()
```

### Filter data
Enter a feature and its required max and/or min values to generate a DataFrame that meets these specifications.

```python
nga.df_filtered = nga.filter_data(feature = "nuclear_area", 
                                  min_value = 30, # Enter value (float) or None
                                  max_value = 300, # Enter value (float) or None
                                  df = nga.df_filtered # If None is given, then uses raw data (nga.df_raw)
                                 )
```

### Find number of DNA Dots
This is an additional nuclear feature, whose measurement takes time. Thus, it is better to measure it on the filtered data. This feature finds the number of high intensity DNA dots in each nucleus. The output is the same DataFrame given as input but with an extra-column for **DNA dots**.

```python
nga.df_filtered = nga.find_dna_dots(df = nga.df_filtered, 
                                    box_size = 10)
```

### Calculate Spatial Entropy
This is an additional nuclear feature, whose measurement takes time. Thus, it is better to measure it on the filtered data. This feature calculates the spatial entropy for each nucleus using the Leibovici method. The output is the same DataFrame given as input but with an extra-column for **Spatial Entropy**.

```python
nga.df_filtered = nga.spatial_entropy(df = nga.df_filtered, 
                                      d = 5)
```

### Save data as a CSV file

```python
nga.toCSV(df = nga.df_filtered, filename = "filtered_output.csv")
```
